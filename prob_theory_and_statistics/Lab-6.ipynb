```python
T = 4
L = 0.1
h = 0.01
C = 0.8
C2 = C*C
du = 0.0015
```

```python
ts = np.arange(0, T + h, h)
xs = np.arange(0, L + h, h)
```

```python
nx, nt = len(xs), len(ts)
u = np.zeros((nt, nx))
```

```python
"""
for col_pos in range(1, len(index) - 1):
    u[0][col_pos] = sin(columns[col_pos])
"""
for j, x in enumerate(xs):
    if x <= L / 2:
        u[0][j] = 2 * du * x / L
    else:
        u[0][j] = 2 * du * (1 - x / L)
       
```

```python
for j in range(1, nx - 1):
    u[1][j] = u[0][j] + C2/2 * (u[0][j+1] - 2 *u[0][j] + u[0][j-1])
```

```python
for i in range(2, nt):
        for j in range(1, nx - 1):
            u[i][j] = 2 * u[i - 1][j] - u[i - 2][j] + C2 * (u[i - 1][j + 1] - 2 * u[i - 1][j] + u[i - 1][j - 1])
```
